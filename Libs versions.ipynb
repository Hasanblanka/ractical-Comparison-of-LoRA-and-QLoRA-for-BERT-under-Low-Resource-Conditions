```python
import importlib, importlib.metadata as md, platform
import pandas as pd, torch

pkgs = ["transformers","datasets","peft","bitsandbytes","accelerate",
        "evaluate","seqeval","numpy","pandas","torch","tqdm","os"]

rows = []
for p in pkgs:
    try:
        v = md.version(p)
    except Exception:
        try:
            m = importlib.import_module(p)
            v = getattr(m, "__version__", getattr(m, "VERSION", "unknown"))
        except Exception:
            v = "NOT INSTALLED"
    rows.append((p, str(v)))

rows += [("python", platform.python_version()),
         ("platform", platform.platform()),
         ("torch_cuda_available", str(torch.cuda.is_available())),
         ("torch_cuda_device_count", str(torch.cuda.device_count() if torch.cuda.is_available() else 0))]

df = pd.DataFrame(rows, columns=["package_or_item","version"])

```
